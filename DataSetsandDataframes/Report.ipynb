```python
# Import pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame with balance.txt
file_path = 'balance.txt'
data = pd.read_csv(file_path, sep=' ')
data_df = pd.DataFrame(data)
data_df.head()
```

```python
# Compare the average income based on ethnicity

groupped_by_ethnicity = data_df.groupby('Ethnicity')['Income'].mean()


x = groupped_by_ethnicity.index
y = groupped_by_ethnicity.values

plt.bar(x,y,color = 'orange', width= 0.3,  edgecolor='black')
plt.xlabel('Ethnicity')
plt.ylabel('Average Income')
plt.title('Average Income by Ethnicity')
plt.show()

highest_group = groupped_by_ethnicity.max()

print(f"The average income by ethnicity is {groupped_by_ethnicity}")
```

```python
grouoped_by_married = data_df.groupby('Married')['Balance'].mean()
grouoped_by_married

x = grouoped_by_married.index
y = grouoped_by_married.values

plt.bar(x,y, color = 'b', width= 0.2, edgecolor='black')
plt.xlabel('Married')
plt.ylabel('Average Balance')
plt.title('Average Balance by Marital Status')
plt.show()

highest_group = grouoped_by_married.max()

print(f"The average balance by marital status: {grouoped_by_married} \n")
```

On average, do married or single people have a higher Balance?

Looking at the average Balance between married and singles as indicated in the bar chart and the computed average, there is a marginal difference between married and single with singles having a higher balance of 13.49.

```python
# What is the highest income in our dataset?

highest_income = data_df['Income'].max()

print(f"The highest income is {highest_income}")
```

```python
# What is the lowest in our dataset?

lowest_income = data_df['Income'].min()

print(f"The lowest income is {lowest_income}")
```

```python
# How many cards do we have recorded in our dataset?

total_cards_recorded = data_df['Cards'].sum()

print(f"The total cards recorded in our datasets = {total_cards_recorded}")
```

```python
# How many females do we have information for vs how many males?


males_vs_females_info = data_df['Gender'].value_counts()


males_vs_females_info.plot(kind='bar', color='b', width= 0.2, edgecolor='black')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Males vs Females')
plt.show()


# Number of females we have information for vs numnber of males
males_vs_females_info = data_df.groupby('Gender').count()

males_vs_females_info

```
