```python
from sklearn.datasets import load_diabetes
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data_df = load_diabetes(as_frame=True).data[['s1', 's2']]
data_df.head()
```

```python
data_df = data_df.rename(columns={'s1': 'total_serum', 's2': 'lipoproteins'})
data_df.head()
```

```python
plt.figure()
sns.scatterplot(data=data_df, x = 'total_serum', y='lipoproteins', color= 'orange')
plt.show()
plt.close()
```

```python
# To allow compatibility, we must ensure that our model inputs are in the shape (n_samples, n_features)

X = data_df['total_serum'].values.reshape(-1,1)
y = data_df['lipoproteins'].values
```

```python
# We now fit and predict our model

linear_model = LinearRegression()
linear_model.fit(X,y)

```

```python
# Plot model and data
y_predicted_model = linear_model.predict(X)
plt.figure()
sns.scatterplot(data=data_df, x = 'total_serum', y='lipoproteins', color= 'orange')
plt.plot(X, y_predicted_model)
plt.show()
plt.close()
```

```python
predicted_serum = [[0.20]]

x_predicted = np.append(X, predicted_serum).reshape(-1,1)
y_predicted = linear_model.predict(x_predicted)

plt.figure()
sns.scatterplot(data=data_df, x = 'total_serum', y='lipoproteins', color= 'orange')
plt.plot(x_predicted, y_predicted, color = 'r')
plt.show()
plt.close()
```
